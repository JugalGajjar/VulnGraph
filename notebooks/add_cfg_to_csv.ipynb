```python
import os
import subprocess
import json
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/csv/cleaned_data.csv")
```

```python
df.head()
```

```python
df["label"].value_counts()
```

```python
for idx, row in df.iterrows():
    try:
        # Write code to file
        java_file = f"../data/raw/{row["id"]}.java"
        with open(java_file, "w", encoding="utf-8") as f:
            f.write(row["code"])
        
        # Run PROGEX command
        cmd = [
            "java", "-jar",
            "../src/progex_parser/progex-v3.4.5/progex.jar",
            "-cfg", "-lang", "java", "-format", "json",
            "-outdir", "../data/graphs", java_file
        ]
        subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    except Exception as e:
        print(f"Error at id {row["id"]}: {e}")
```

```python
cfgs ={}
```

```python
for idx, row in df.iterrows():
    try:
        # Read the generated JSON
        json_file = f"../data/graphs/{row['id']}-CFG.json"
        if os.path.exists(json_file):
            with open(json_file, "r", encoding="utf-8") as jf:
                cfg_data = json.load(jf)
        else:
            cfg_data = None
        cfgs[row["id"]] = cfg_data
    except Exception as e:
        print(f"Error at id {row['id']}: {e}")
        cfgs[row["id"]] = None
```

```python
df.head()
```

```python
column_position = df.columns.get_loc("label")
df.insert(column_position, "cfg", df["id"].map(cfgs))
```
